```python
import matplotlib.pyplot as plt


def check_point(xmin,xmax,ymin,ymax,x1,y1):
    if(x1>=xmin and x1<=xmax):
        if(y1>=ymin and y1<=ymax):
            return True
        else:
            return False        
    else:
        return False
    
def calc_point(xmin,ymin,xmax,ymax,x1,x2,y1,y2):
    m=(y2-y1)/(x2-x1)
    if(check_point(xmin,xmax,ymin,ymax,x1,y1) and check_point(xmin,xmax,ymin,ymax,x2,y2)):
        print("Line is Completely Inside the Required Region :- Nothing to Clip")
    elif(not check_point(xmin,xmax,ymin,ymax,x1,y1)) and (not check_point(xmin,xmax,ymin,ymax,x2,y2)):
        print("Line is Completely Outside the Required Region :- Nothing to Clip")
    else:
        if check_point(xmin,xmax,ymin,ymax,x1,y1):
            a=x1
            b=y1
            c=x2
            d=y2
        else:
            a=x2
            b=y2
            c=x1
            d=y1
        x=[]
        y=[]
        x.append(a)
        y.append(b)
        if c<=xmin:
            x.append(xmin)
            y.append(int(b+(m*(xmin-a))))
            
        elif c>=xmax:
            x.append(xmax)
            y.append(int(b+(m*(xmax-a))))
            
        elif d<=ymin:
            y.append(ymin)
            x.append(int(a+((ymin-b)/m)))
            
        else :
            y.append(ymax)
            x.append(int(a+((ymax-b)/m)))
    
        plt.ylim(ymin,ymax)
        plt.xlim(xmin,xmax)
        plt.title("Cohen and Sutherland Line Clipping Algorithm")
        plt.plot(x,y)

xmin=int(input("Enter Lower Point X Cordinate of region :- "))
ymin=int(input("Enter Lower Point Y Cordinate of region :- "))
xmax=int(input("Enter Upper Point X Cordinate of region :- "))
ymax=int(input("Enter Upper Point Y Cordinate of region :- "))

x1=int(input("Enter  X Cordinate of 1st Point of line :- "))
y1=int(input("Enter  Y Cordinate of 1st Point of line :- "))
x2=int(input("Enter  X Cordinate of 2nd Point of line :- "))
y2=int(input("Enter  Y Cordinate of 2nd Point of line :- "))
calc_point(xmin,ymin,xmax,ymax,x1,x2,y1,y2)
```
