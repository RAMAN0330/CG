```python
import matplotlib.pyplot as plt



def calc_point(x1,x2,y1,y2):
    dy=y2-y1
    dx=x2-x1
    m=dy/dx
    p=(2*dy)-dx
    x=[]
    y=[]
    xnext=x1
    ynext=y1
    x.append(xnext)
    y.append(ynext)
    while (xnext<x2 or ynext<y2):
        if m<1:
            xnext=xnext+1
            x.append(xnext)
            if p<0:
                y.append(ynext)
                p=p+(2*dy)
            else:
                ynext=ynext+1
                y.append(ynext)
                p=p+2*(dy-dx)
        else:
            ynext=ynext+1
            y.append(ynext)
            if p<0:
                xnext=xnext
                x.append(xnext)
                p=p+(2*dx)
            else:
                xnext=xnext+1
                x.append(xnext)
                p=p+2*(dx-dy)
    print("X list :- ",x)
    print("Y list :- ",y)
    drawpoints(x,y)
    
def drawpoints(x,y):
    plt.plot(x,y)
    plt.title("Bresenham's Line Drawing Algorithm ")
    plt.xlabel(" X-Axis ")
    plt.ylabel(" Y-Axis ")
    plt.show()
    

x1=int(input("Enter Starting Point X Cordinate :- "))
y1=int(input("Enter Starting Point Y Cordinate :- "))
x2=int(input("Enter Ending Point X Cordinate :- "))
y2=int(input("Enter Ending Point Y Cordinate :- "))
calc_point(x1,x2,y1,y2)
```
